```python
import numpy as np
import pandas as pd
import scipy as sp
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn import tree


#import data
demo=pd.read_csv('demographic.csv')
diet=pd.read_csv('diet.csv')
exam=pd.read_csv('examination.csv')
labs=pd.read_csv('labs.csv')
med=pd.read_csv('medications_3.csv')
qns=pd.read_csv('questionnaire.csv')
var=pd.read_csv('Variables_of_importance.csv')

#merging data
demo.loc[demo['RIDAGEYR']>=19,'Age']=demo['RIDAGEYR']
demo.loc[demo['RIDAGEYR']<19,'Age']=demo['RIDEXAGM']/12
#create a candidate tables with all datas merged
data_merged=demo
data_merged=data_merged.merge(labs)
data_merged=data_merged.merge(diet)
data_merged=data_merged.merge(qns)
data_merged=data_merged.merge(exam)

#find columns that we want
#creating data_merged of impt variables
col1=var['Code'].tolist()
col2=data_merged.columns.tolist()

imax=len(col1)-1
jmax=len(col2)-1
col=['SEQN','Age']
for i in range(0,imax):
        for j in range(0,jmax):
                if col1[i]==col2[j]:
                    col.append(col1[i])
                    break
                    
                    
data_merged_impt=data_merged[col]


#merge med data together with data_merged
#keywords=['Elevated blood glucose level',
 #         'Gestational diabetes mellitus',
  #        'Prevent diabetes',
   #       'Prevent diabetic kidney disease',
    #      'Type 1 diabetes mellitus',
     #     'Type 2 diabetes mellitus',]

imax=len(data_merged['SEQN'])
druglist=med['RXDDRUG']
illnesslist=med['RXDRSD1']
med_SEQN=med['SEQN']
SEQN=data_merged['SEQN']

table1=[]

for i in range(0,imax):
    lis1=druglist.loc[med['SEQN']==SEQN[i]]
    lis2=illnesslist.loc[med['SEQN']==SEQN[i]].drop_duplicates(keep='first')
    data_merged_impt.loc[i,['druglist']]=lis1.str.cat(sep=',')
    data_merged_impt.loc[i,['illness']]=lis2.str.cat(sep=',')

print(data_merged_impt.head())


#X=data_merged['LBXGH']#Glycohemoglobin (%)
#Y=data_merged['BMXBMI']#BMI
#plt.scatter(X,Y)
#plt.show()
```
